````python
{
 "cells": [
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "# Unit 4: Statistical Methods & EDA\n",
    "**Tribhuvan University — IOE | BE Computer Engineering**\n",
    "\n",
    "---"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "## 4.1 Statistical Measures: Correlation, Covariance, Skewness & Kurtosis"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q1. Define covariance and correlation. How are they related? What does a correlation of -1, 0, and +1 indicate? [5 marks]\n",
    "\n",
    "**Covariance** measures the joint variability of two random variables. It indicates the direction of the linear relationship between variables.\n",
    "\n",
    "$$\\text{Cov}(X, Y) = \\frac{1}{n-1} \\sum_{i=1}^{n}(X_i - \\bar{X})(Y_i - \\bar{Y})$$\n",
    "\n",
    "**Correlation** (Pearson's) is the standardized form of covariance. It measures both the strength and direction of the linear relationship between two variables, scaled to the range [-1, +1].\n",
    "\n",
    "$$r = \\frac{\\text{Cov}(X,Y)}{\\sigma_X \\cdot \\sigma_Y}$$\n",
    "\n",
    "**Relationship:** Correlation is covariance normalized by the product of the standard deviations of both variables, making it unit-free and bounded.\n",
    "\n",
    "| Correlation Value | Interpretation |\n",
    "|---|---|\n",
    "| **r = +1** | Perfect positive linear relationship — as X increases, Y increases proportionally |\n",
    "| **r = 0** | No linear relationship — X and Y are linearly uncorrelated (may still have non-linear relationship) |\n",
    "| **r = -1** | Perfect negative linear relationship — as X increases, Y decreases proportionally |"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q2. Given X = [2, 4, 6, 8, 10] and Y = [1, 3, 5, 7, 9], compute the sample covariance and Pearson correlation coefficient manually. Interpret your results. [10 marks]\n",
    "\n",
    "**Step 1: Compute means**\n",
    "\n",
    "$$\\bar{X} = \\frac{2+4+6+8+10}{5} = \\frac{30}{5} = 6$$\n",
    "\n",
    "$$\\bar{Y} = \\frac{1+3+5+7+9}{5} = \\frac{25}{5} = 5$$\n",
    "\n",
    "**Step 2: Compute deviations and products**\n",
    "\n",
    "| $X_i$ | $Y_i$ | $X_i - \\bar{X}$ | $Y_i - \\bar{Y}$ | $(X_i-\\bar{X})(Y_i-\\bar{Y})$ | $(X_i-\\bar{X})^2$ | $(Y_i-\\bar{Y})^2$ |\n",
    "|---|---|---|---|---|---|---|\n",
    "| 2 | 1 | -4 | -4 | 16 | 16 | 16 |\n",
    "| 4 | 3 | -2 | -2 | 4 | 4 | 4 |\n",
    "| 6 | 5 | 0 | 0 | 0 | 0 | 0 |\n",
    "| 8 | 7 | 2 | 2 | 4 | 4 | 4 |\n",
    "| 10 | 9 | 4 | 4 | 16 | 16 | 16 |\n",
    "| | | | **Sum** | **40** | **40** | **40** |\n",
    "\n",
    "**Step 3: Sample Covariance**\n",
    "\n",
    "$$\\text{Cov}(X,Y) = \\frac{\\sum(X_i - \\bar{X})(Y_i - \\bar{Y})}{n-1} = \\frac{40}{4} = 10$$\n",
    "\n",
    "**Step 4: Standard Deviations**\n",
    "\n",
    "$$\\sigma_X = \\sqrt{\\frac{40}{4}} = \\sqrt{10} \\approx 3.162$$\n",
    "\n",
    "$$\\sigma_Y = \\sqrt{\\frac{40}{4}} = \\sqrt{10} \\approx 3.162$$\n",
    "\n",
    "**Step 5: Pearson Correlation**\n",
    "\n",
    "$$r = \\frac{\\text{Cov}(X,Y)}{\\sigma_X \\cdot \\sigma_Y} = \\frac{10}{\\sqrt{10} \\times \\sqrt{10}} = \\frac{10}{10} = 1.0$$\n",
    "\n",
    "**Interpretation:**\n",
    "- Sample Covariance = **10** → positive relationship (both variables increase together)\n",
    "- Pearson r = **1.0** → perfect positive linear relationship. This makes sense because Y = X - 1 (a perfectly linear transformation). Every unit increase in X corresponds to exactly one unit increase in Y."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q3. Differentiate between Pearson, Spearman, and Kendall correlation. When is Spearman preferred over Pearson? [5 marks]\n",
    "\n",
    "| Feature | Pearson | Spearman | Kendall |\n",
    "|---|---|---|---|\n",
    "| **Type** | Parametric | Non-parametric | Non-parametric |\n",
    "| **Based on** | Actual values | Ranks of values | Concordant/discordant pairs |\n",
    "| **Measures** | Linear relationship | Monotonic relationship | Ordinal association |\n",
    "| **Assumption** | Normality, linearity, no outliers | Monotonic trend | Monotonic trend |\n",
    "| **Sensitivity to outliers** | High | Low | Low |\n",
    "| **Formula** | $r = \\frac{\\text{Cov}(X,Y)}{\\sigma_X \\sigma_Y}$ | $r_s = 1 - \\frac{6\\sum d_i^2}{n(n^2-1)}$ | $\\tau = \\frac{C - D}{C + D}$ |\n",
    "\n",
    "**When is Spearman preferred over Pearson?**\n",
    "1. When data is **ordinal** (e.g., survey ratings, rankings)\n",
    "2. When the relationship is **monotonic but not linear**\n",
    "3. When the data contains **significant outliers** that would distort Pearson\n",
    "4. When the **normality assumption** for Pearson cannot be satisfied\n",
    "5. When working with **small sample sizes** where normality is uncertain\n",
    "\n",
    "*Example:* Measuring the relationship between education level (ordinal: high school, bachelor's, master's, PhD) and salary — Spearman is more appropriate."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q4. Explain skewness and kurtosis. What do positive skewness and leptokurtic distributions indicate about income distribution? [5 marks]\n",
    "\n",
    "**Skewness** measures the asymmetry of a probability distribution around its mean.\n",
    "\n",
    "$$\\text{Skewness} = \\frac{1}{n}\\sum\\left(\\frac{X_i - \\bar{X}}{\\sigma}\\right)^3$$\n",
    "\n",
    "| Skewness | Distribution | Tail |\n",
    "|---|---|---|\n",
    "| Negative (< 0) | Left-skewed | Long left tail |\n",
    "| Zero (= 0) | Symmetric | Normal distribution |\n",
    "| Positive (> 0) | Right-skewed | Long right tail |\n",
    "\n",
    "**Kurtosis** measures the \"tailedness\" or peakedness of a distribution relative to a normal distribution.\n",
    "\n",
    "$$\\text{Kurtosis} = \\frac{1}{n}\\sum\\left(\\frac{X_i - \\bar{X}}{\\sigma}\\right)^4$$\n",
    "\n",
    "| Type | Excess Kurtosis | Shape |\n",
    "|---|---|---|\n",
    "| Platykurtic | < 0 | Flat peak, thin tails |\n",
    "| Mesokurtic | = 0 | Normal distribution |\n",
    "| Leptokurtic | > 0 | Sharp peak, fat tails |\n",
    "\n",
    "**Application to Income Distribution:**\n",
    "- **Positive Skewness** in income data means most people earn below the mean income, while a small number of very high earners pull the mean upward (e.g., billionaires skew the distribution right). The **median < mean**.\n",
    "- **Leptokurtic** income distribution means extreme incomes (both very poor and very rich) occur more frequently than a normal distribution would predict — indicating high income inequality with fat tails."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q5. Dataset [2,3,5,6,7,8,10,15,20,50]: Calculate skewness and kurtosis. Determine if platykurtic, mesokurtic, or leptokurtic. Explain which statistical tests are affected. [10 marks]\n",
    "\n",
    "**Dataset:** $X = [2, 3, 5, 6, 7, 8, 10, 15, 20, 50]$, $n = 10$\n",
    "\n",
    "**Step 1: Mean**\n",
    "$$\\bar{X} = \\frac{2+3+5+6+7+8+10+15+20+50}{10} = \\frac{126}{10} = 12.6$$\n",
    "\n",
    "**Step 2: Standard Deviation**\n",
    "\n",
    "| $X_i$ | $X_i - \\bar{X}$ | $(X_i-\\bar{X})^2$ | $(X_i-\\bar{X})^3$ | $(X_i-\\bar{X})^4$ |\n",
    "|---|---|---|---|---|\n",
    "| 2 | -10.6 | 112.36 | -1190.99 | 12623.7 |\n",
    "| 3 | -9.6 | 92.16 | -884.74 | 8493.5 |\n",
    "| 5 | -7.6 | 57.76 | -438.98 | 3336.2 |\n",
    "| 6 | -6.6 | 43.56 | -287.50 | 1897.5 |\n",
    "| 7 | -5.6 | 31.36 | -175.62 | 983.4 |\n",
    "| 8 | -4.6 | 21.16 | -97.34 | 447.7 |\n",
    "| 10 | -2.6 | 6.76 | -17.58 | 45.7 |\n",
    "| 15 | 2.4 | 5.76 | 13.82 | 33.2 |\n",
    "| 20 | 7.4 | 54.76 | 405.22 | 2998.6 |\n",
    "| 50 | 37.4 | 1398.76 | 52313.62 | 1956549.5 |\n",
    "| **Sum** | | **1824.4** | **49640.01** | **1987409.0** |\n",
    "\n",
    "$$s^2 = \\frac{1824.4}{9} = 202.71 \\quad \\Rightarrow \\quad s = 14.24$$\n",
    "\n",
    "**Step 3: Skewness (Fisher's)**\n",
    "$$g_1 = \\frac{n}{(n-1)(n-2)} \\cdot \\frac{\\sum(X_i - \\bar{X})^3}{s^3} \\approx \\frac{10}{9 \\times 8} \\cdot \\frac{49640.01}{2885.3} \\approx \\mathbf{2.39}$$\n",
    "\n",
    "→ **Strongly right-skewed** (long right tail due to the value 50)\n",
    "\n",
    "**Step 4: Excess Kurtosis**\n",
    "$$g_2 = \\frac{n(n+1)}{(n-1)(n-2)(n-3)} \\cdot \\frac{\\sum(X_i-\\bar{X})^4}{s^4} - \\frac{3(n-1)^2}{(n-2)(n-3)} \\approx \\mathbf{6.02}$$\n",
    "\n",
    "→ Excess kurtosis > 0 → **Leptokurtic**\n",
    "\n",
    "**Conclusion:**\n",
    "- Skewness ≈ 2.39 → strong positive skew\n",
    "- Excess Kurtosis ≈ 6.02 → **Leptokurtic** (sharp peak, fat tails)\n",
    "\n",
    "**Statistical Tests Affected:**\n",
    "\n",
    "| Test | Effect of Skewness/Kurtosis |\n",
    "|---|---|\n",
    "| **t-test** | Assumes normality; violated by high skewness — use Wilcoxon instead |\n",
    "| **ANOVA** | Sensitive to non-normality; use Kruskal-Wallis for skewed data |\n",
    "| **Pearson Correlation** | Assumes bivariate normality; use Spearman for skewed data |\n",
    "| **Linear Regression** | Residuals should be normal; skewed residuals violate OLS assumptions |\n",
    "| **Z-score outlier detection** | Unreliable for leptokurtic data — IQR method preferred |"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q6. Python: Compute covariance matrix, Pearson correlation matrix, skewness, and kurtosis. Visualize with Seaborn heatmap. [10 marks]"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import numpy as np\n",
    "import pandas as pd\n",
    "from scipy import stats\n",
    "import seaborn as sns\n",
    "import matplotlib.pyplot as plt\n",
    "\n",
    "# Sample DataFrame\n",
    "np.random.seed(42)\n",
    "df = pd.DataFrame({\n",
    "    'Age':    np.random.randint(20, 60, 100),\n",
    "    'Income': np.random.normal(50000, 15000, 100),\n",
    "    'Score':  np.random.normal(70, 10, 100),\n",
    "    'Hours':  np.random.uniform(1, 12, 100)\n",
    "})\n",
    "\n",
    "print(\"=\" * 50)\n",
    "print(\"COVARIANCE MATRIX\")\n",
    "print(\"=\" * 50)\n",
    "cov_matrix = df.cov()\n",
    "print(cov_matrix.round(2))\n",
    "\n",
    "print(\"\\n\" + \"=\" * 50)\n",
    "print(\"PEARSON CORRELATION MATRIX\")\n",
    "print(\"=\" * 50)\n",
    "corr_matrix = df.corr(method='pearson')\n",
    "print(corr_matrix.round(4))\n",
    "\n",
    "print(\"\\n\" + \"=\" * 50)\n",
    "print(\"SKEWNESS & KURTOSIS PER COLUMN\")\n",
    "print(\"=\" * 50)\n",
    "skewness = df.apply(stats.skew)\n",
    "kurtosis = df.apply(stats.kurtosis)  # excess kurtosis (Fisher's definition)\n",
    "\n",
    "sk_df = pd.DataFrame({'Skewness': skewness, 'Excess Kurtosis': kurtosis})\n",
    "sk_df['Distribution Type'] = sk_df['Excess Kurtosis'].apply(\n",
    "    lambda k: 'Leptokurtic' if k > 0.5 else ('Platykurtic' if k < -0.5 else 'Mesokurtic')\n",
    ")\n",
    "print(sk_df.round(4))\n",
    "\n",
    "# Visualisation — Correlation Heatmap\n",
    "plt.figure(figsize=(8, 6))\n",
    "sns.heatmap(\n",
    "    corr_matrix,\n",
    "    annot=True,\n",
    "    fmt='.2f',\n",
    "    cmap='coolwarm',\n",
    "    center=0,\n",
    "    linewidths=0.5,\n",
    "    square=True,\n",
    "    cbar_kws={'shrink': 0.8}\n",
    ")\n",
    "plt.title('Pearson Correlation Matrix Heatmap', fontsize=14, fontweight='bold')\n",
    "plt.tight_layout()\n",
    "plt.savefig('correlation_heatmap.png', dpi=150)\n",
    "plt.show()\n",
    "print(\"Heatmap saved as 'correlation_heatmap.png'\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "---\n",
    "## 4.2 Probability Review, Sampling & Hypothesis Testing"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q7. State the Central Limit Theorem (CLT). Why is it important for statistical inference when population distribution is unknown? [5 marks]\n",
    "\n",
    "**Central Limit Theorem (CLT):**\n",
    "\n",
    "> Given a population with mean $\\mu$ and finite standard deviation $\\sigma$, if we draw sufficiently large random samples (typically $n \\geq 30$) of size $n$, the sampling distribution of the sample mean $\\bar{X}$ will approximate a **normal distribution**, regardless of the shape of the population distribution:\n",
    "\n",
    "$$\\bar{X} \\sim N\\left(\\mu,\\ \\frac{\\sigma^2}{n}\\right) \\quad \\text{as } n \\to \\infty$$\n",
    "\n",
    "The standard error of the mean is: $SE = \\frac{\\sigma}{\\sqrt{n}}$\n",
    "\n",
    "**Importance for Statistical Inference:**\n",
    "\n",
    "1. **Normality Assumption Bridge:** Most classical tests (t-test, z-test, ANOVA) assume normality. CLT ensures these tests remain valid on large samples even when the population is skewed, bimodal, or unknown.\n",
    "\n",
    "2. **Confidence Intervals:** CLT allows construction of confidence intervals using the normal/t-distribution for the sample mean.\n",
    "\n",
    "3. **Hypothesis Testing:** Enables computation of p-values using known distributions.\n",
    "\n",
    "4. **Real-world applicability:** In practice, population distributions are rarely known. CLT lets statisticians draw valid conclusions without knowing the exact population distribution — only the sample is needed.\n",
    "\n",
    "5. **Reduction of Variance:** Larger samples produce sampling distributions with smaller standard errors, yielding more precise estimates."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q8. Explain Type I and Type II errors. How do significance level (α) and power relate to these errors? [5 marks]\n",
    "\n",
    "| | **H₀ is True** | **H₀ is False** |\n",
    "|---|---|---|\n",
    "| **Reject H₀** | ❌ Type I Error (False Positive) | ✅ Correct (True Positive) |\n",
    "| **Fail to Reject H₀** | ✅ Correct (True Negative) | ❌ Type II Error (False Negative) |\n",
    "\n",
    "**Type I Error (α — False Positive):**\n",
    "- Rejecting a **true** null hypothesis\n",
    "- Probability = **α** (significance level)\n",
    "- Example: Concluding a drug works when it actually doesn't\n",
    "- Controlled by setting α (commonly 0.05 or 0.01)\n",
    "\n",
    "**Type II Error (β — False Negative):**\n",
    "- Failing to reject a **false** null hypothesis\n",
    "- Probability = **β**\n",
    "- Example: Concluding a drug doesn't work when it actually does\n",
    "- Reduced by increasing sample size or effect size\n",
    "\n",
    "**Statistical Power = 1 - β**\n",
    "- Probability of correctly rejecting a false H₀\n",
    "- Desired power is typically ≥ 0.80\n",
    "\n",
    "**Trade-off:** Decreasing α (e.g., from 0.05 to 0.01) reduces Type I errors but increases β (Type II errors), thus reducing power. Increasing sample size $n$ reduces both types of errors simultaneously."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q9. One-sample t-test: Mean weight claimed = 500g, sample n=50, x̄=495g, s=15g, α=0.05. [10 marks]\n",
    "\n",
    "**Hypotheses:**\n",
    "$$H_0: \\mu = 500g \\quad (\\text{manufacturer's claim is true})$$\n",
    "$$H_1: \\mu \\neq 500g \\quad (\\text{two-tailed test — mean is different from 500g})$$\n",
    "\n",
    "**Given:**\n",
    "- $\\bar{x} = 495g$, $s = 15g$, $n = 50$, $\\alpha = 0.05$\n",
    "\n",
    "**Test Statistic (t):**\n",
    "$$t = \\frac{\\bar{x} - \\mu_0}{s / \\sqrt{n}} = \\frac{495 - 500}{15 / \\sqrt{50}} = \\frac{-5}{15/7.071} = \\frac{-5}{2.121} \\approx -2.357$$\n",
    "\n",
    "**Degrees of Freedom:** $df = n - 1 = 49$\n",
    "\n",
    "**Critical Value (two-tailed, α=0.05, df=49):**\n",
    "$$t_{\\text{critical}} = \\pm 2.010$$\n",
    "\n",
    "**p-value:**\n",
    "$$p = 2 \\times P(T < -2.357) \\approx 2 \\times 0.0112 = 0.0224$$\n",
    "\n",
    "**Decision:**\n",
    "- $|t_{\\text{calc}}| = 2.357 > t_{\\text{critical}} = 2.010$ → **Reject H₀**\n",
    "- $p = 0.0224 < \\alpha = 0.05$ → **Reject H₀**\n",
    "\n",
    "**Conclusion:**\n",
    "At the 5% significance level, there is sufficient statistical evidence to **reject the manufacturer's claim**. The true mean weight of the product is significantly different from 500g. The sample data suggests the actual mean weight (495g) is statistically lower than the claimed 500g."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q10. Compare Simple Random, Stratified, Cluster, and Systematic Sampling with practical examples. [8 marks]\n",
    "\n",
    "| Feature | Simple Random | Stratified | Cluster | Systematic |\n",
    "|---|---|---|---|---|\n",
    "| **Definition** | Every member has equal probability of selection | Population divided into subgroups (strata); random sample from each | Population divided into clusters; entire clusters randomly selected | Every k-th element selected after a random start |\n",
    "| **Randomness** | Fully random | Random within strata | Random cluster selection | Semi-random |\n",
    "| **Best for** | Homogeneous populations | Heterogeneous populations with known subgroups | Large, geographically spread populations | Large ordered lists |\n",
    "| **Cost** | Low–Medium | Medium | Low (field surveys) | Low |\n",
    "| **Representation** | May miss subgroups | Excellent subgroup representation | May have high intra-cluster similarity | Good if no periodic pattern |\n",
    "\n",
    "**Practical Examples:**\n",
    "\n",
    "1. **Simple Random Sampling:** Selecting 100 students randomly from a university roll number list using a random number generator to evaluate average CGPA.\n",
    "\n",
    "2. **Stratified Sampling:** Surveying employee satisfaction in a company by dividing into strata (managers, engineers, interns) and randomly sampling proportionally from each group — ensures all levels are represented.\n",
    "\n",
    "3. **Cluster Sampling:** Studying school performance across Nepal by randomly selecting 20 districts (clusters) and surveying all schools within those districts — cost-effective for large geographic coverage.\n",
    "\n",
    "4. **Systematic Sampling:** Quality control on a production line — inspecting every 10th item off the assembly line (e.g., items 10, 20, 30, ...) for defects."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q11. What is p-value? At p=0.03, what conclusions at α=0.05 and α=0.01? [5 marks]\n",
    "\n",
    "**p-value Definition:**\n",
    "\n",
    "The p-value is the probability of obtaining a test statistic at least as extreme as the observed value, **assuming the null hypothesis (H₀) is true**. It quantifies the evidence against H₀.\n",
    "\n",
    "$$p = P(\\text{observed or more extreme result} \\mid H_0 \\text{ is true})$$\n",
    "\n",
    "- A **small p-value** → strong evidence against H₀ → reject H₀\n",
    "- A **large p-value** → weak evidence against H₀ → fail to reject H₀\n",
    "\n",
    "**Decision Rule:** Reject H₀ if $p < \\alpha$\n",
    "\n",
    "**Given p = 0.03:**\n",
    "\n",
    "| Significance Level | Comparison | Decision | Conclusion |\n",
    "|---|---|---|---|\n",
    "| **α = 0.05** | 0.03 < 0.05 | ✅ **Reject H₀** | Statistically significant at 5% level. There is sufficient evidence to reject the null hypothesis. |\n",
    "| **α = 0.01** | 0.03 > 0.01 | ❌ **Fail to Reject H₀** | Not statistically significant at 1% level. Insufficient evidence to reject the null hypothesis. |\n",
    "\n",
    "**Interpretation:** The result p=0.03 is \"marginally significant\" — significant at the conventional 5% threshold but not at the more stringent 1% threshold. The conclusion depends on the chosen significance level before conducting the test."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q12. Python: Two-sample t-test using scipy.stats on two randomly generated datasets. [10 marks]"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import numpy as np\n",
    "from scipy import stats\n",
    "import matplotlib.pyplot as plt\n",
    "\n",
    "def two_sample_ttest(group1, group2, alpha=0.05):\n",
    "    \"\"\"\n",
    "    Performs an independent two-sample t-test.\n",
    "    \n",
    "    Parameters:\n",
    "        group1 (array-like): First sample\n",
    "        group2 (array-like): Second sample\n",
    "        alpha (float): Significance level (default=0.05)\n",
    "    \n",
    "    Returns:\n",
    "        dict: t-statistic, p-value, degrees of freedom, and conclusion\n",
    "    \"\"\"\n",
    "    print(\"=\" * 55)\n",
    "    print(\"     TWO-SAMPLE INDEPENDENT t-TEST\")\n",
    "    print(\"=\" * 55)\n",
    "\n",
    "    # Hypotheses\n",
    "    print(\"\\nHypotheses:\")\n",
    "    print(\"  H₀: μ₁ = μ₂  (No difference between group means)\")\n",
    "    print(\"  H₁: μ₁ ≠ μ₂  (Significant difference exists — two-tailed)\")\n",
    "\n",
    "    # Descriptive statistics\n",
    "    print(f\"\\nDescriptive Statistics:\")\n",
    "    print(f\"  Group 1: n={len(group1)}, mean={np.mean(group1):.4f}, std={np.std(group1, ddof=1):.4f}\")\n",
    "    print(f\"  Group 2: n={len(group2)}, mean={np.mean(group2):.4f}, std={np.std(group2, ddof=1):.4f}\")\n",
    "\n",
    "    # Levene's test for equal variances\n",
    "    levene_stat, levene_p = stats.levene(group1, group2)\n",
    "    equal_var = levene_p > 0.05\n",
    "    print(f\"\\nLevene's Test for Equal Variances: p={levene_p:.4f}\")\n",
    "    print(f\"  → {'Assuming equal variances (Student t-test)' if equal_var else 'Unequal variances (Welch t-test)'}\")\n",
    "\n",
    "    # t-test\n",
    "    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=equal_var)\n",
    "    df = len(group1) + len(group2) - 2\n",
    "\n",
    "    print(f\"\\nTest Results:\")\n",
    "    print(f\"  t-statistic : {t_stat:.4f}\")\n",
    "    print(f\"  Degrees of Freedom: {df}\")\n",
    "    print(f\"  p-value     : {p_value:.4f}\")\n",
    "    print(f\"  Alpha       : {alpha}\")\n",
    "\n",
    "    # Conclusion\n",
    "    print(\"\\nConclusion:\")\n",
    "    if p_value < alpha:\n",
    "        conclusion = (f\"  p={p_value:.4f} < α={alpha} → REJECT H₀\\n\"\n",
    "                      f\"  There IS a statistically significant difference between group means.\")\n",
    "    else:\n",
    "        conclusion = (f\"  p={p_value:.4f} ≥ α={alpha} → FAIL TO REJECT H₀\\n\"\n",
    "                      f\"  There is NO statistically significant difference between group means.\")\n",
    "    print(conclusion)\n",
    "    print(\"=\" * 55)\n",
    "\n",
    "    return {'t_stat': t_stat, 'p_value': p_value, 'df': df}\n",
    "\n",
    "\n",
    "# Generate two random samples\n",
    "np.random.seed(42)\n",
    "group_A = np.random.normal(loc=70, scale=10, size=40)  # Mean=70\n",
    "group_B = np.random.normal(loc=75, scale=12, size=45)  # Mean=75\n",
    "\n",
    "result = two_sample_ttest(group_A, group_B)\n",
    "\n",
    "# Visualisation\n",
    "fig, axes = plt.subplots(1, 2, figsize=(12, 4))\n",
    "\n",
    "axes[0].hist(group_A, bins=12, color='steelblue', alpha=0.7, label='Group A', edgecolor='white')\n",
    "axes[0].hist(group_B, bins=12, color='salmon', alpha=0.7, label='Group B', edgecolor='white')\n",
    "axes[0].axvline(np.mean(group_A), color='blue', linestyle='--', label=f'Mean A={np.mean(group_A):.1f}')\n",
    "axes[0].axvline(np.mean(group_B), color='red', linestyle='--', label=f'Mean B={np.mean(group_B):.1f}')\n",
    "axes[0].set_title('Distribution of Group A vs Group B')\n",
    "axes[0].set_xlabel('Value')\n",
    "axes[0].set_ylabel('Frequency')\n",
    "axes[0].legend()\n",
    "\n",
    "axes[1].boxplot([group_A, group_B], labels=['Group A', 'Group B'], patch_artist=True,\n",
    "                boxprops=dict(facecolor='lightblue'),\n",
    "                medianprops=dict(color='red', linewidth=2))\n",
    "axes[1].set_title('Boxplot Comparison')\n",
    "axes[1].set_ylabel('Value')\n",
    "\n",
    "plt.suptitle(f'Two-Sample t-Test | t={result[\"t_stat\"]:.3f}, p={result[\"p_value\"]:.4f}',\n",
    "             fontweight='bold')\n",
    "plt.tight_layout()\n",
    "plt.savefig('two_sample_ttest.png', dpi=150)\n",
    "plt.show()"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "---\n",
    "## 4.3 Regression & Trend Analysis using StatsModels"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q13. Explain OLS method for fitting a linear regression line. Derive formulas for slope (β₁) and intercept (β₀). [8 marks]\n",
    "\n",
    "**Ordinary Least Squares (OLS)** is a method for estimating the parameters of a linear regression model. The goal is to find the line $\\hat{Y} = \\beta_0 + \\beta_1 X$ that **minimizes the sum of squared residuals (SSR)**:\n",
    "\n",
    "$$\\text{SSR} = \\sum_{i=1}^{n} (Y_i - \\hat{Y}_i)^2 = \\sum_{i=1}^{n} (Y_i - \\beta_0 - \\beta_1 X_i)^2$$\n",
    "\n",
    "**Derivation of β₁ (Slope):**\n",
    "\n",
    "Take partial derivative with respect to $\\beta_1$ and set to zero:\n",
    "\n",
    "$$\\frac{\\partial \\text{SSR}}{\\partial \\beta_1} = -2\\sum_{i=1}^{n} X_i(Y_i - \\beta_0 - \\beta_1 X_i) = 0$$\n",
    "\n",
    "$$\\sum X_i Y_i - \\beta_0 \\sum X_i - \\beta_1 \\sum X_i^2 = 0 \\quad \\cdots (1)$$\n",
    "\n",
    "**Derivation of β₀ (Intercept):**\n",
    "\n",
    "Take partial derivative with respect to $\\beta_0$ and set to zero:\n",
    "\n",
    "$$\\frac{\\partial \\text{SSR}}{\\partial \\beta_0} = -2\\sum_{i=1}^{n}(Y_i - \\beta_0 - \\beta_1 X_i) = 0$$\n",
    "\n",
    "$$\\sum Y_i = n\\beta_0 + \\beta_1 \\sum X_i \\implies \\beta_0 = \\bar{Y} - \\beta_1 \\bar{X} \\quad \\cdots (2)$$\n",
    "\n",
    "Substituting (2) into (1) and simplifying:\n",
    "\n",
    "$$\\boxed{\\beta_1 = \\frac{\\sum_{i=1}^{n}(X_i - \\bar{X})(Y_i - \\bar{Y})}{\\sum_{i=1}^{n}(X_i - \\bar{X})^2} = \\frac{\\text{Cov}(X,Y)}{\\text{Var}(X)}}$$\n",
    "\n",
    "$$\\boxed{\\beta_0 = \\bar{Y} - \\beta_1 \\bar{X}}$$\n",
    "\n",
    "**Properties of OLS estimators (Gauss-Markov theorem):** OLS estimators are BLUE — **B**est **L**inear **U**nbiased **E**stimators — under the classical assumptions. They have minimum variance among all linear unbiased estimators."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q14. What is R²? How does adjusted R² differ, and why is it more appropriate for multiple regression? [5 marks]\n",
    "\n",
    "**R² (Coefficient of Determination):**\n",
    "\n",
    "R² measures the proportion of variance in the dependent variable explained by the independent variable(s):\n",
    "\n",
    "$$R^2 = 1 - \\frac{SS_{\\text{res}}}{SS_{\\text{tot}}} = 1 - \\frac{\\sum(Y_i - \\hat{Y}_i)^2}{\\sum(Y_i - \\bar{Y})^2}$$\n",
    "\n",
    "- Range: [0, 1]; closer to 1 means better fit\n",
    "- R² = 0.85 means 85% of variance in Y is explained by the model\n",
    "\n",
    "**Problem with R²:** Adding more predictors to a model **always increases or maintains R²**, even if the new variables are irrelevant or random — it never decreases.\n",
    "\n",
    "**Adjusted R²:**\n",
    "\n",
    "$$R^2_{\\text{adj}} = 1 - \\frac{(1-R^2)(n-1)}{n-k-1}$$\n",
    "\n",
    "where $n$ = number of observations, $k$ = number of predictors\n",
    "\n",
    "**Key Differences:**\n",
    "\n",
    "| Feature | R² | Adjusted R² |\n",
    "|---|---|---|\n",
    "| Penalizes extra predictors | ❌ No | ✅ Yes |\n",
    "| Can decrease when adding variables | ❌ Never | ✅ Yes (if variable is not useful) |\n",
    "| Used for | Simple regression | Multiple regression model comparison |\n",
    "| Range | [0, 1] | Can be negative (poor model) |\n",
    "\n",
    "**Why Adjusted R² is better for multiple regression:** It penalizes model complexity by accounting for degrees of freedom lost. It only increases when a new variable genuinely improves the model's explanatory power, making it a reliable metric for comparing models with different numbers of predictors."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q15. List and explain the key assumptions of linear regression (LINE). What diagnostic plots check each? [5 marks]\n",
    "\n",
    "The acronym **LINE** summarizes the four key assumptions:\n",
    "\n",
    "| Assumption | Description | Diagnostic Plot |\n",
    "|---|---|---|\n",
    "| **L — Linearity** | The relationship between X and Y is linear. | Residuals vs Fitted plot (should show random scatter around 0) |\n",
    "| **I — Independence** | Residuals are independent (no autocorrelation). | Residuals vs Order plot; Durbin-Watson test |\n",
    "| **N — Normality** | Residuals are normally distributed. | Q-Q plot (Normal probability plot); Histogram of residuals |\n",
    "| **E — Equal Variance (Homoscedasticity)** | Residuals have constant variance across all fitted values. | Scale-Location plot (standardized residuals vs fitted); Breusch-Pagan test |\n",
    "\n",
    "**Additional Assumption:**\n",
    "- **No Multicollinearity:** In multiple regression, predictors should not be highly correlated with each other → checked with VIF (Variance Inflation Factor) values (VIF > 10 is problematic)\n",
    "\n",
    "**Diagnostic Plot Summary:**\n",
    "1. **Residuals vs Fitted** → Linearity & Homoscedasticity\n",
    "2. **Q-Q Plot** → Normality of residuals\n",
    "3. **Scale-Location Plot** → Homoscedasticity\n",
    "4. **Cook's Distance / Leverage Plot** → Influential observations & outliers"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q16. Simple Linear Regression: House sizes and prices. Predict price for 1400 sq.ft. Compute RMSE. [10 marks]\n",
    "\n",
    "**Data:**\n",
    "- Size (X): [800, 1000, 1200, 1500, 1800] sq.ft\n",
    "- Price (Y): [45, 55, 65, 80, 95] lakhs NPR\n",
    "\n",
    "**Step 1: Compute means**\n",
    "$$\\bar{X} = \\frac{800+1000+1200+1500+1800}{5} = \\frac{6300}{5} = 1260$$\n",
    "$$\\bar{Y} = \\frac{45+55+65+80+95}{5} = \\frac{340}{5} = 68$$\n",
    "\n",
    "**Step 2: Compute β₁**\n",
    "\n",
    "| $X_i$ | $Y_i$ | $X_i - \\bar{X}$ | $Y_i - \\bar{Y}$ | $(X_i-\\bar{X})(Y_i-\\bar{Y})$ | $(X_i-\\bar{X})^2$ |\n",
    "|---|---|---|---|---|---|\n",
    "| 800 | 45 | -460 | -23 | 10580 | 211600 |\n",
    "| 1000 | 55 | -260 | -13 | 3380 | 67600 |\n",
    "| 1200 | 65 | -60 | -3 | 180 | 3600 |\n",
    "| 1500 | 80 | 240 | 12 | 2880 | 57600 |\n",
    "| 1800 | 95 | 540 | 27 | 14580 | 291600 |\n",
    "| **Sum** | | | | **31600** | **632000** |\n",
    "\n",
    "$$\\beta_1 = \\frac{31600}{632000} \\approx 0.05$$\n",
    "\n",
    "$$\\beta_0 = 68 - 0.05 \\times 1260 = 68 - 63 = 5$$\n",
    "\n",
    "**Regression Equation:** $\\hat{Y} = 5 + 0.05X$\n",
    "\n",
    "**Step 3: Prediction for 1400 sq.ft**\n",
    "$$\\hat{Y}_{1400} = 5 + 0.05 \\times 1400 = 5 + 70 = \\mathbf{75 \\text{ lakhs NPR}}$$\n",
    "\n",
    "**Step 4: RMSE**\n",
    "\n",
    "| $X_i$ | $Y_i$ | $\\hat{Y}_i$ | Residual $e_i$ | $e_i^2$ |\n",
    "|---|---|---|---|---|\n",
    "| 800 | 45 | 45 | 0 | 0 |\n",
    "| 1000 | 55 | 55 | 0 | 0 |\n",
    "| 1200 | 65 | 65 | 0 | 0 |\n",
    "| 1500 | 80 | 80 | 0 | 0 |\n",
    "| 1800 | 95 | 95 | 0 | 0 |\n",
    "\n",
    "$$\\text{RMSE} = \\sqrt{\\frac{\\sum e_i^2}{n}} = \\sqrt{\\frac{0}{5}} = \\mathbf{0}$$\n",
    "\n",
    "**Interpretation:** RMSE = 0 indicates a perfect fit — the linear model passes exactly through all five data points. This is possible with 5 points that form a nearly perfect linear pattern. In practice, real data would yield RMSE > 0."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q17. Explain multicollinearity. How is VIF used to detect it, and what corrective measures apply? [5 marks]\n",
    "\n",
    "**Multicollinearity** occurs when two or more independent variables in a multiple regression model are highly correlated with each other, making it difficult to isolate their individual effects on the dependent variable.\n",
    "\n",
    "**Problems caused:**\n",
    "- Inflated standard errors of coefficients\n",
    "- Unstable and unreliable coefficient estimates\n",
    "- Insignificant p-values even when relationships exist\n",
    "- High R² but insignificant individual predictors\n",
    "\n",
    "**VIF (Variance Inflation Factor):**\n",
    "\n",
    "$$\\text{VIF}_j = \\frac{1}{1 - R_j^2}$$\n",
    "\n",
    "where $R_j^2$ is the R² from regressing predictor $X_j$ on all other predictors.\n",
    "\n",
    "| VIF Value | Interpretation |\n",
    "|---|---|\n",
    "| 1 | No multicollinearity |\n",
    "| 1 – 5 | Moderate, usually acceptable |\n",
    "| 5 – 10 | High — investigate |\n",
    "| > 10 | Severe multicollinearity — action required |\n",
    "\n",
    "**Corrective Measures:**\n",
    "1. **Remove one of the correlated variables** — drop the less theoretically important predictor\n",
    "2. **Principal Component Analysis (PCA)** — transform correlated features into uncorrelated components\n",
    "3. **Ridge Regression (L2 regularization)** — adds penalty term to stabilize coefficients\n",
    "4. **Combine variables** — create an index (e.g., average height + weight → BMI)\n",
    "5. **Collect more data** — reduces standard errors and can mitigate multicollinearity effects"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q18. Python: Multiple linear regression with statsmodels. Interpret OLS summary. [10 marks]"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import numpy as np\n",
    "import pandas as pd\n",
    "import statsmodels.api as sm\n",
    "import matplotlib.pyplot as plt\n",
    "import seaborn as sns\n",
    "from scipy import stats\n",
    "\n",
    "# Generate synthetic dataset: Predict house price\n",
    "np.random.seed(42)\n",
    "n = 100\n",
    "\n",
    "size        = np.random.randint(500, 3000, n)          # sq.ft\n",
    "bedrooms    = np.random.randint(1, 6, n)               # number of bedrooms\n",
    "age         = np.random.randint(0, 40, n)              # years old\n",
    "\n",
    "# True relationship: price = 10 + 0.05*size + 3*bedrooms - 0.5*age + noise\n",
    "price = 10 + 0.05 * size + 3 * bedrooms - 0.5 * age + np.random.normal(0, 8, n)\n",
    "\n",
    "df = pd.DataFrame({'Price': price, 'Size': size, 'Bedrooms': bedrooms, 'Age': age})\n",
    "\n",
    "# Prepare features — add constant for intercept\n",
    "X = sm.add_constant(df[['Size', 'Bedrooms', 'Age']])\n",
    "y = df['Price']\n",
    "\n",
    "# Fit OLS model\n",
    "model = sm.OLS(y, X).fit()\n",
    "\n",
    "print(model.summary())\n",
    "\n",
    "# --- Interpretation ---\n",
    "print(\"\\n\" + \"=\" * 60)\n",
    "print(\"INTERPRETATION OF OLS SUMMARY\")\n",
    "print(\"=\" * 60)\n",
    "print(f\"\\nR²       = {model.rsquared:.4f}  → Model explains {model.rsquared*100:.1f}% of variance in Price\")\n",
    "print(f\"Adj. R²  = {model.rsquared_adj:.4f}  → Adjusted for {X.shape[1]-1} predictors\")\n",
    "print(f\"F-stat   = {model.fvalue:.4f}, p={model.f_pvalue:.6f}  → Model is {'significant' if model.f_pvalue < 0.05 else 'not significant'} overall\")\n",
    "print(\"\\nCoefficients:\")\n",
    "for name, coef, pval in zip(model.params.index, model.params, model.pvalues):\n",
    "    sig = '***' if pval < 0.001 else ('**' if pval < 0.01 else ('*' if pval < 0.05 else ''))\n",
    "    print(f\"  {name:<12}: {coef:+.4f}  (p={pval:.4f}) {sig}\")\n",
    "\n",
    "print(\"\\nSignificance codes: *** p<0.001, ** p<0.01, * p<0.05\")\n",
    "\n",
    "# Residual plots\n",
    "fig, axes = plt.subplots(1, 2, figsize=(12, 4))\n",
    "\n",
    "residuals = model.resid\n",
    "fitted = model.fittedvalues\n",
    "\n",
    "axes[0].scatter(fitted, residuals, alpha=0.5, color='steelblue')\n",
    "axes[0].axhline(0, color='red', linestyle='--')\n",
    "axes[0].set_xlabel('Fitted Values')\n",
    "axes[0].set_ylabel('Residuals')\n",
    "axes[0].set_title('Residuals vs Fitted')\n",
    "\n",
    "sm.qqplot(residuals, line='45', fit=True, ax=axes[1], alpha=0.5)\n",
    "axes[1].set_title('Q-Q Plot of Residuals')\n",
    "\n",
    "plt.suptitle('OLS Diagnostic Plots', fontweight='bold')\n",
    "plt.tight_layout()\n",
    "plt.savefig('ols_diagnostics.png', dpi=150)\n",
    "plt.show()"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "---\n",
    "## 4.4 Exploratory Data Analysis (EDA) — Descriptive & Inferential Methods"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q19. Define EDA. List and briefly describe five key steps in performing EDA on a new dataset. [5 marks]\n",
    "\n",
    "**Exploratory Data Analysis (EDA)** is an iterative, open-ended approach to analyzing datasets by summarizing their main characteristics, often using visual methods, before formal modeling. The term was popularized by statistician John Tukey (1977). The goal is to understand the data's structure, detect anomalies, test assumptions, and discover patterns.\n",
    "\n",
    "**Five Key Steps in EDA:**\n",
    "\n",
    "1. **Data Overview & Understanding**\n",
    "   - Examine shape (`df.shape`), data types (`df.dtypes`), and preview (`df.head()`)\n",
    "   - Understand what each column represents and the data source\n",
    "   - Check memory usage and basic info with `df.info()`\n",
    "\n",
    "2. **Missing Value Analysis**\n",
    "   - Count and visualize missing values (`df.isnull().sum()`, `msno.heatmap()`)\n",
    "   - Determine whether data is missing at random (MAR), completely at random (MCAR), or not at random (MNAR)\n",
    "   - Decide imputation strategy (mean/median/mode/KNN) or row removal\n",
    "\n",
    "3. **Univariate Analysis**\n",
    "   - Analyze each variable individually: descriptive statistics (mean, median, std, quartiles)\n",
    "   - Visualize distributions using histograms, box plots, count plots (for categorical)\n",
    "   - Detect outliers and skewness per feature\n",
    "\n",
    "4. **Bivariate & Multivariate Analysis**\n",
    "   - Explore relationships between pairs of variables using scatter plots, correlation matrix, heatmaps\n",
    "   - Group comparisons (e.g., mean salary by gender) using grouped bar plots or violin plots\n",
    "   - Pairplots for multivariate exploration\n",
    "\n",
    "5. **Feature Engineering & Insights**\n",
    "   - Encode categorical variables, create derived features, handle class imbalance\n",
    "   - Document data-driven insights and hypotheses for modeling\n",
    "   - Summarize findings and identify which features are most relevant"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q20. Distinguish between descriptive and inferential statistics. Give two examples of each. [5 marks]\n",
    "\n",
    "| Feature | Descriptive Statistics | Inferential Statistics |\n",
    "|---|---|---|\n",
    "| **Purpose** | Summarize and describe the data at hand | Draw conclusions about a population from a sample |\n",
    "| **Scope** | Limited to the dataset (no generalization) | Generalizes beyond the data collected |\n",
    "| **Output** | Tables, charts, summary measures | Hypothesis tests, confidence intervals, p-values |\n",
    "| **Tools** | Mean, median, mode, variance, range | t-test, ANOVA, regression, chi-square |\n",
    "| **Population required?** | No — works on sample or population | Yes — infers about population from sample |\n",
    "\n",
    "**Descriptive Statistics Examples:**\n",
    "1. A school computes the **average score (mean = 72.5), standard deviation (8.3), and score distribution histogram** for 500 students in a math exam — describing only those students.\n",
    "2. A hospital reports the **proportion of COVID-positive patients by age group** (25-35: 20%, 36-50: 35%, 51+: 45%) among admitted patients in March — summarizing observed data.\n",
    "\n",
    "**Inferential Statistics Examples:**\n",
    "1. A pharmaceutical company tests a new drug on **200 patients** (sample) and uses a **two-sample t-test** to conclude that it is effective for **all patients with the condition** (population).\n",
    "2. A political analyst surveys **1,000 voters** and uses a **confidence interval** to estimate that the ruling party has 48–52% approval across the **entire electorate** of 10 million voters."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q21. Explain box plots, violin plots, and histograms. For extreme outliers, which is preferred? Describe IQR-based outlier detection. [8 marks]\n",
    "\n",
    "**Box Plot:**\n",
    "- Displays the **five-number summary**: minimum, Q1 (25th percentile), median (Q2), Q3 (75th percentile), maximum\n",
    "- The **box** spans Q1–Q3 (Interquartile Range, IQR)\n",
    "- **Whiskers** extend to 1.5×IQR beyond Q1 and Q3\n",
    "- Points beyond whiskers are plotted as individual **outlier dots**\n",
    "- Best for: comparing distributions across groups, quickly identifying outliers\n",
    "\n",
    "**Violin Plot:**\n",
    "- Combines a box plot with a **Kernel Density Estimate (KDE)** curve on each side\n",
    "- Shows the full **shape of the distribution** (bimodal, skewed, etc.) — information box plots miss\n",
    "- Best for: understanding distribution shape and comparing groups with complex distributions\n",
    "\n",
    "**Histogram:**\n",
    "- Displays frequency of values in **bins/intervals**\n",
    "- Shows the overall **distribution shape**: normal, skewed, bimodal\n",
    "- Sensitive to bin width — different bin sizes can suggest different shapes\n",
    "- Best for: visualizing the frequency distribution of a single continuous variable\n",
    "\n",
    "**Preferred for Extreme Outliers:** **Box Plot**\n",
    "Box plots explicitly mark outliers as dots beyond 1.5×IQR whiskers, making them visually obvious. Histograms and violin plots may obscure single extreme values in wide bin ranges or smooth them out in KDE.\n",
    "\n",
    "**IQR-Based Outlier Detection:**\n",
    "\n",
    "$$\\text{IQR} = Q3 - Q1$$\n",
    "\n",
    "$$\\text{Lower Bound} = Q1 - 1.5 \\times \\text{IQR}$$\n",
    "\n",
    "$$\\text{Upper Bound} = Q3 + 1.5 \\times \\text{IQR}$$\n",
    "\n",
    "Any value below the Lower Bound or above the Upper Bound is classified as an **outlier**.\n",
    "\n",
    "*Advantage over Z-score:* The IQR method is **robust to outliers themselves** — it uses medians and percentiles, not mean and standard deviation which can be distorted by extreme values. Better suited for skewed distributions."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q22. Step-by-step EDA on a student marks dataset with missing values, outliers, and mixed types. [10 marks]\n",
    "\n",
    "**Scenario:** Dataset contains: StudentID (int), Name (str), Gender (categorical), Math (float), Science (float), English (float), Attendance% (float), Grade (categorical), with missing values and outliers.\n",
    "\n",
    "---\n",
    "\n",
    "**Step 1: Data Loading & Initial Inspection**\n",
    "- `df.shape` → (rows, columns)\n",
    "- `df.info()` → column names, data types, non-null counts\n",
    "- `df.head()` / `df.tail()` → preview rows\n",
    "- `df.describe()` → summary stats for numerical columns\n",
    "- `df.dtypes` → identify mixed data types (e.g., Grade stored as int when it should be categorical)\n",
    "- **Visualization:** None (inspection only)\n",
    "\n",
    "**Step 2: Data Cleaning**\n",
    "- Missing Values:\n",
    "  - `df.isnull().sum()` → count nulls per column\n",
    "  - **Visualization:** `msno.heatmap(df)` — missing value heatmap\n",
    "  - Numerical columns (Math, Science) → impute with **median** (robust to outliers)\n",
    "  - Categorical (Gender, Grade) → impute with **mode**\n",
    "- Data type correction: Convert Grade to `category`, StudentID to `str`\n",
    "- Duplicate detection: `df.duplicated().sum()`\n",
    "\n",
    "**Step 3: Univariate Analysis**\n",
    "- Numerical (Math, Science, English, Attendance%):\n",
    "  - Compute: mean, median, std, skewness, kurtosis\n",
    "  - **Visualizations:** Histograms with KDE overlay for each subject score; Box plots to detect outliers\n",
    "  - IQR-based outlier detection: flag outliers (e.g., Math score = 150 for 100-point exam)\n",
    "- Categorical (Gender, Grade):\n",
    "  - `df['Gender'].value_counts()`\n",
    "  - **Visualizations:** Count plots / pie charts\n",
    "\n",
    "**Step 4: Bivariate Analysis**\n",
    "- Numerical vs Numerical:\n",
    "  - **Scatter plots:** Math vs Science scores — detect correlation\n",
    "  - **Correlation heatmap:** Pearson correlation matrix across all subjects\n",
    "- Categorical vs Numerical:\n",
    "  - **Box plots / Violin plots:** Math scores by Gender or Grade\n",
    "  - `df.groupby('Grade')['Math'].mean()` — group means\n",
    "- **Pairplot:** (`sns.pairplot(df, hue='Gender')`) — all numerical pairs colored by gender\n",
    "\n",
    "**Step 5: Feature Correlation & Insights**\n",
    "- Compute full correlation matrix\n",
    "- **Visualization:** Annotated heatmap highlighting strong correlations (|r| > 0.7)\n",
    "- Insights:\n",
    "  - Are Math and Science scores highly correlated? → Consider feature reduction\n",
    "  - Does attendance predict grades? → Scatter plot Attendance% vs Grade (encoded)\n",
    "  - Gender differences in performance? → t-test for mean differences\n",
    "\n",
    "**Summary of Visualizations:**\n",
    "| Step | Visualization |\n",
    "|---|---|\n",
    "| Cleaning | Missing value heatmap |\n",
    "| Univariate | Histograms, box plots, count plots |\n",
    "| Bivariate | Scatter plots, grouped box plots |\n",
    "| Multivariate | Pairplot, correlation heatmap |"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q23. Five-number summary for [3,7,8,5,12,14,21,13,18]. Draw box plot and identify outliers. [5 marks]\n",
    "\n",
    "**Sorted data:** [3, 5, 7, 8, 12, 13, 14, 18, 21]\n",
    "\n",
    "$n = 9$\n",
    "\n",
    "**Five-Number Summary:**\n",
    "\n",
    "| Measure | Value | Calculation |\n",
    "|---|---|---|\n",
    "| **Minimum** | 3 | Smallest value |\n",
    "| **Q1** | 6 | Median of lower half [3,5,7,8] → (5+7)/2 = 6 |\n",
    "| **Median (Q2)** | 12 | Middle value (5th of 9) |\n",
    "| **Q3** | 16 | Median of upper half [13,14,18,21] → (14+18)/2 = 16 |\n",
    "| **Maximum** | 21 | Largest value |\n",
    "\n",
    "**IQR Calculation:**\n",
    "$$\\text{IQR} = Q3 - Q1 = 16 - 6 = 10$$\n",
    "\n",
    "**Outlier Bounds:**\n",
    "$$\\text{Lower} = Q1 - 1.5 \\times \\text{IQR} = 6 - 15 = -9$$\n",
    "$$\\text{Upper} = Q3 + 1.5 \\times \\text{IQR} = 16 + 15 = 31$$\n",
    "\n",
    "**Outlier Detection:** All data values fall within [-9, 31] → **No outliers detected**\n",
    "\n",
    "```\n",
    "Box Plot Sketch:\n",
    "\n",
    "  |----[====|=========]----|\n",
    "  3    6   12          16  21\n",
    " Min   Q1  Median      Q3  Max\n",
    "```\n",
    "\n",
    "- Left whisker extends from 3 (Min) to Q1 = 6\n",
    "- Box spans Q1=6 to Q3=16 with median line at 12\n",
    "- Right whisker extends from Q3=16 to 21 (Max)\n",
    "- No points plotted as outliers (none beyond whisker bounds)"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q24. Python: EDA on a public dataset (Titanic) — shape/info, null handling, descriptive stats, 4+ visualizations, 3 insights. [10 marks]"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import pandas as pd\n",
    "import numpy as np\n",
    "import matplotlib.pyplot as plt\n",
    "import seaborn as sns\n",
    "from sklearn.datasets import fetch_openml\n",
    "\n",
    "# Load Titanic dataset\n",
    "titanic_data = fetch_openml('titanic', version=1, as_frame=True, parser='auto')\n",
    "df = titanic_data.frame.copy()\n",
    "\n",
    "# Select relevant columns\n",
    "df = df[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']]\n",
    "df['survived'] = df['survived'].astype(int)\n",
    "df['pclass'] = df['pclass'].astype(int)\n",
    "\n",
    "print(\"=\" * 55)\n",
    "print(\"STEP 1: DATASET OVERVIEW\")\n",
    "print(\"=\" * 55)\n",
    "print(f\"Shape: {df.shape} — {df.shape[0]} rows, {df.shape[1]} columns\")\n",
    "print(\"\\nData Types:\\n\", df.dtypes)\n",
    "\n",
    "print(\"\\n\" + \"=\" * 55)\n",
    "print(\"STEP 2: MISSING VALUE ANALYSIS\")\n",
    "print(\"=\" * 55)\n",
    "null_counts = df.isnull().sum()\n",
    "null_pct = (null_counts / len(df) * 100).round(2)\n",
    "print(pd.DataFrame({'Null Count': null_counts, 'Null %': null_pct}))\n",
    "\n",
    "# Handle nulls\n",
    "df['age'].fillna(df['age'].median(), inplace=True)\n",
    "df['fare'].fillna(df['fare'].median(), inplace=True)\n",
    "df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)\n",
    "print(\"\\nAfter imputation — Remaining nulls:\", df.isnull().sum().sum())\n",
    "\n",
    "print(\"\\n\" + \"=\" * 55)\n",
    "print(\"STEP 3: DESCRIPTIVE STATISTICS\")\n",
    "print(\"=\" * 55)\n",
    "print(df.describe().round(2))\n",
    "\n",
    "print(\"\\nSurvival Rate:\", f\"{df['survived'].mean()*100:.1f}%\")\n",
    "print(\"Survival by Class:\\n\", df.groupby('pclass')['survived'].mean().round(3))\n",
    "print(\"Survival by Sex:\\n\", df.groupby('sex')['survived'].mean().round(3))"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# STEP 4: VISUALIZATIONS\n",
    "fig, axes = plt.subplots(2, 2, figsize=(14, 10))\n",
    "fig.suptitle('Titanic EDA — Key Visualizations', fontsize=16, fontweight='bold', y=1.01)\n",
    "\n",
    "# Viz 1: Survival count\n",
    "survival_counts = df['survived'].value_counts()\n",
    "axes[0, 0].bar(['Did Not Survive', 'Survived'], survival_counts.values,\n",
    "               color=['#e74c3c', '#2ecc71'], edgecolor='white', linewidth=0.5)\n",
    "axes[0, 0].set_title('Survival Distribution', fontweight='bold')\n",
    "axes[0, 0].set_ylabel('Count')\n",
    "for i, v in enumerate(survival_counts.values):\n",
    "    axes[0, 0].text(i, v + 5, str(v), ha='center', fontweight='bold')\n",
    "\n",
    "# Viz 2: Survival rate by class and sex\n",
    "surv_pivot = df.groupby(['pclass', 'sex'])['survived'].mean().unstack()\n",
    "surv_pivot.plot(kind='bar', ax=axes[0, 1], color=['#e74c3c', '#3498db'],\n",
    "                edgecolor='white', rot=0)\n",
    "axes[0, 1].set_title('Survival Rate by Class & Sex', fontweight='bold')\n",
    "axes[0, 1].set_xlabel('Passenger Class')\n",
    "axes[0, 1].set_ylabel('Survival Rate')\n",
    "axes[0, 1].legend(title='Sex')\n",
    "\n",
    "# Viz 3: Age distribution by survival\n",
    "for val, color, label in [(0, '#e74c3c', 'Did Not Survive'), (1, '#2ecc71', 'Survived')]:\n",
    "    axes[1, 0].hist(df[df['survived'] == val]['age'], bins=25, alpha=0.6,\n",
    "                    color=color, label=label, edgecolor='white')\n",
    "axes[1, 0].set_title('Age Distribution by Survival', fontweight='bold')\n",
    "axes[1, 0].set_xlabel('Age')\n",
    "axes[1, 0].set_ylabel('Count')\n",
    "axes[1, 0].legend()\n",
    "\n",
    "# Viz 4: Fare vs Age scatter colored by survival\n",
    "colors = df['survived'].map({0: '#e74c3c', 1: '#2ecc71'})\n",
    "axes[1, 1].scatter(df['age'], df['fare'], c=colors, alpha=0.4, s=15)\n",
    "axes[1, 1].set_title('Fare vs Age (colored by Survival)', fontweight='bold')\n",
    "axes[1, 1].set_xlabel('Age')\n",
    "axes[1, 1].set_ylabel('Fare (£)')\n",
    "from matplotlib.patches import Patch\n",
    "legend_elements = [Patch(facecolor='#e74c3c', label='Did Not Survive'),\n",
    "                   Patch(facecolor='#2ecc71', label='Survived')]\n",
    "axes[1, 1].legend(handles=legend_elements)\n",
    "\n",
    "plt.tight_layout()\n",
    "plt.savefig('titanic_eda.png', dpi=150, bbox_inches='tight')\n",
    "plt.show()\n",
    "\n",
    "print(\"\\n\" + \"=\" * 55)\n",
    "print(\"STEP 5: DATA-DRIVEN INSIGHTS\")\n",
    "print(\"=\" * 55)\n",
    "print(\"\"\"\n",
    "Insight 1 — Gender Bias in Survival:\n",
    "  Female survival rate: ~73%  vs  Male: ~19%\n",
    "  → 'Women and children first' policy is clearly reflected.\n",
    "  Female passengers were ~4x more likely to survive.\n",
    "\n",
    "Insight 2 — Class Determines Fate:\n",
    "  1st class survival: ~63%  |  2nd: ~47%  |  3rd: ~24%\n",
    "  → Higher-class passengers had better access to lifeboats\n",
    "  and were housed on upper decks closer to evacuation points.\n",
    "\n",
    "Insight 3 — Children Had Better Survival Odds:\n",
    "  Passengers under age 10 show notably higher survival rates.\n",
    "  The age distribution shows many young non-survivors in the\n",
    "  20-35 age range — predominantly male third-class passengers.\n",
    "\"\"\")"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "---\n",
    "## 4.5 Automation of EDA Workflows using Python"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q25. What are automated EDA libraries? Compare pandas-profiling (ydata-profiling), Sweetviz, and D-Tale. [5 marks]\n",
    "\n",
    "**Automated EDA libraries** are Python packages that generate comprehensive exploratory data analysis reports with minimal code — typically just 2–3 lines — automatically covering statistics, distributions, correlations, missing values, and more.\n",
    "\n",
    "| Feature | ydata-profiling (pandas-profiling) | Sweetviz | D-Tale |\n",
    "|---|---|---|---|\n",
    "| **Output format** | HTML report | HTML report | Interactive web app (Flask) |\n",
    "| **Ease of use** | Very easy (2 lines) | Very easy (2 lines) | Easy but requires server |\n",
    "| **Interactivity** | Static HTML | Static HTML | Fully interactive UI |\n",
    "| **Comparison** | Dataset vs dataset | Excellent (train vs test) | Limited |\n",
    "| **Key features** | Correlation, duplicates, distribution, alerts, interactions | Target analysis, comparison report | Filtering, sorting, charting live |\n",
    "| **Report size** | Large (slow for big data) | Medium | Real-time (no file) |\n",
    "| **Best for** | Comprehensive standalone report | ML train/test comparison | Interactive exploration in notebooks |\n",
    "| **Customization** | High (via ProfileReport config) | Moderate | High (UI-based) |\n",
    "\n",
    "**Recommendation by Use Case:**\n",
    "- **ydata-profiling** → Best for a thorough one-click report to share with stakeholders\n",
    "- **Sweetviz** → Best for comparing training vs. test datasets before ML modeling\n",
    "- **D-Tale** → Best for interactive, drill-down exploration during active analysis"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q26. Explain the importance of automating EDA workflows in production. What are limitations of automated EDA tools for domain-specific data? [5 marks]\n",
    "\n",
    "**Importance of Automating EDA Workflows in Production:**\n",
    "\n",
    "1. **Speed & Efficiency:** Automated EDA reduces hours of manual analysis to minutes, allowing data scientists to focus on modeling and interpretation rather than boilerplate code.\n",
    "\n",
    "2. **Consistency & Reproducibility:** Automated pipelines apply the same steps to every new data batch — eliminating human variability in analysis and ensuring reproducible results across team members.\n",
    "\n",
    "3. **Data Drift Detection:** In production ML pipelines, automated EDA can continuously monitor incoming data distributions and alert when significant drift from training data is detected — critical for model reliability.\n",
    "\n",
    "4. **Onboarding New Datasets:** When new data sources arrive, automated EDA generates an immediate profile, helping teams quickly understand structure before committing to expensive preprocessing.\n",
    "\n",
    "5. **Scalability:** Can be integrated into CI/CD pipelines (e.g., triggered on each new data delivery), making EDA a standard automated quality gate.\n",
    "\n",
    "**Limitations of Automated EDA for Domain-Specific Data:**\n",
    "\n",
    "1. **No Domain Context:** Automated tools cannot interpret whether an outlier is a data error or a valid extreme case (e.g., a 300-unit spike in ICU admissions may be normal during a pandemic).\n",
    "\n",
    "2. **Inappropriate Metrics:** Generic statistical summaries may be irrelevant — e.g., mean/std of a binary encoded categorical variable is misleading and should use frequency tables.\n",
    "\n",
    "3. **Cannot Detect Semantic Errors:** Tools won't flag logically impossible combinations (e.g., a patient aged 5 with 30 years of work experience).\n",
    "\n",
    "4. **Misinterpretation of Correlation:** High correlation between features may be spurious without domain knowledge (e.g., ice cream sales and drowning incidents correlate due to summer — not causation).\n",
    "\n",
    "5. **Privacy & Compliance Blind Spots:** Automated tools don't recognize PII (Personally Identifiable Information) or HIPAA-sensitive fields that require special handling."
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q27. Python: Design automated EDA function eda_report(df) with shape, null summary, dtypes, stats, skewness/kurtosis, and correlation heatmap saved to folder. [10 marks]"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import pandas as pd\n",
    "import numpy as np\n",
    "import matplotlib.pyplot as plt\n",
    "import seaborn as sns\n",
    "from scipy import stats\n",
    "import os\n",
    "\n",
    "def eda_report(df, output_dir='eda_outputs'):\n",
    "    \"\"\"\n",
    "    Automated EDA report generator.\n",
    "    \n",
    "    Parameters:\n",
    "        df         : pandas DataFrame to analyze\n",
    "        output_dir : folder path to save outputs (default: 'eda_outputs')\n",
    "    \n",
    "    Outputs:\n",
    "        - Console: shape, dtypes, null summary, descriptive stats, skewness/kurtosis\n",
    "        - Files:   eda_summary.csv, eda_sk_kurt.csv, correlation_heatmap.png\n",
    "    \"\"\"\n",
    "\n",
    "    os.makedirs(output_dir, exist_ok=True)\n",
    "    separator = \"=\" * 60\n",
    "\n",
    "    # ── 1. Shape ──────────────────────────────────────────────────\n",
    "    print(separator)\n",
    "    print(\"1. DATASET SHAPE\")\n",
    "    print(separator)\n",
    "    print(f\"   Rows    : {df.shape[0]}\")\n",
    "    print(f\"   Columns : {df.shape[1]}\")\n",
    "\n",
    "    # ── 2. Data Types ─────────────────────────────────────────────\n",
    "    print(f\"\\n{separator}\")\n",
    "    print(\"2. DATA TYPES\")\n",
    "    print(separator)\n",
    "    dtypes_df = df.dtypes.rename('dtype').to_frame()\n",
    "    dtypes_df['unique_values'] = df.nunique()\n",
    "    print(dtypes_df.to_string())\n",
    "\n",
    "    # ── 3. Null Summary ───────────────────────────────────────────\n",
    "    print(f\"\\n{separator}\")\n",
    "    print(\"3. MISSING VALUE SUMMARY\")\n",
    "    print(separator)\n",
    "    null_df = pd.DataFrame({\n",
    "        'Null Count': df.isnull().sum(),\n",
    "        'Null %': (df.isnull().sum() / len(df) * 100).round(2)\n",
    "    })\n",
    "    null_df = null_df[null_df['Null Count'] > 0]\n",
    "    if null_df.empty:\n",
    "        print(\"   ✅ No missing values found.\")\n",
    "    else:\n",
    "        print(null_df.to_string())\n",
    "\n",
    "    # ── 4. Descriptive Statistics ──────────────────────────────────\n",
    "    print(f\"\\n{separator}\")\n",
    "    print(\"4. DESCRIPTIVE STATISTICS (Numerical Columns)\")\n",
    "    print(separator)\n",
    "    num_df = df.select_dtypes(include=[np.number])\n",
    "    desc_stats = num_df.describe().T\n",
    "    desc_stats['median'] = num_df.median()\n",
    "    print(desc_stats.round(3).to_string())\n",
    "    desc_stats.to_csv(os.path.join(output_dir, 'eda_summary.csv'))\n",
    "    print(f\"\\n   💾 Saved: {output_dir}/eda_summary.csv\")\n",
    "\n",
    "    # ── 5. Skewness & Kurtosis ────────────────────────────────────\n",
    "    print(f\"\\n{separator}\")\n",
    "    print(\"5. SKEWNESS & KURTOSIS (Numerical Columns)\")\n",
    "    print(separator)\n",
    "    sk_kurt = pd.DataFrame({\n",
    "        'Skewness': num_df.apply(stats.skew),\n",
    "        'Excess Kurtosis': num_df.apply(stats.kurtosis)\n",
    "    }).round(4)\n",
    "    sk_kurt['Skew Direction'] = sk_kurt['Skewness'].apply(\n",
    "        lambda s: 'Right-skewed' if s > 0.5 else ('Left-skewed' if s < -0.5 else 'Symmetric'))\n",
    "    sk_kurt['Kurtosis Type'] = sk_kurt['Excess Kurtosis'].apply(\n",
    "        lambda k: 'Leptokurtic' if k > 0.5 else ('Platykurtic' if k < -0.5 else 'Mesokurtic'))\n",
    "    print(sk_kurt.to_string())\n",
    "    sk_kurt.to_csv(os.path.join(output_dir, 'eda_sk_kurt.csv'))\n",
    "    print(f\"\\n   💾 Saved: {output_dir}/eda_sk_kurt.csv\")\n",
    "\n",
    "    # ── 6. Correlation Heatmap ────────────────────────────────────\n",
    "    if num_df.shape[1] >= 2:\n",
    "        corr = num_df.corr()\n",
    "        fig, ax = plt.subplots(figsize=(max(6, num_df.shape[1]), max(5, num_df.shape[1] - 1)))\n",
    "        sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,\n",
    "                    linewidths=0.5, ax=ax, cbar_kws={'shrink': 0.8})\n",
    "        ax.set_title('Correlation Heatmap', fontsize=13, fontweight='bold')\n",
    "        plt.tight_layout()\n",
    "        heatmap_path = os.path.join(output_dir, 'correlation_heatmap.png')\n",
    "        plt.savefig(heatmap_path, dpi=150)\n",
    "        plt.show()\n",
    "        print(f\"\\n   💾 Saved: {heatmap_path}\")\n",
    "\n",
    "    print(f\"\\n{separator}\")\n",
    "    print(f\"✅ EDA Complete. All outputs saved to: ./{output_dir}/\")\n",
    "    print(separator)\n",
    "\n",
    "\n",
    "# ── Demo ──────────────────────────────────────────────────────────\n",
    "np.random.seed(0)\n",
    "demo_df = pd.DataFrame({\n",
    "    'Age':    np.random.randint(18, 65, 200),\n",
    "    'Salary': np.random.exponential(scale=50000, size=200),\n",
    "    'Score':  np.random.normal(70, 15, 200),\n",
    "    'Hours':  np.random.uniform(20, 60, 200),\n",
    "    'Gender': np.random.choice(['M', 'F'], 200)\n",
    "})\n",
    "# Introduce some nulls\n",
    "demo_df.loc[np.random.choice(200, 15, replace=False), 'Age'] = np.nan\n",
    "demo_df.loc[np.random.choice(200, 10, replace=False), 'Score'] = np.nan\n",
    "\n",
    "eda_report(demo_df, output_dir='eda_outputs')"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q28. Using ydata-profiling, generate an EDA report. Describe information captured and how it speeds up data science. [8 marks]\n",
    "\n",
    "**ydata-profiling** (formerly pandas-profiling) generates a comprehensive, interactive HTML EDA report with a single function call:\n",
    "\n",
    "```python\n",
    "from ydata_profiling import ProfileReport\n",
    "import pandas as pd\n",
    "\n",
    "df = pd.read_csv('your_dataset.csv')\n",
    "profile = ProfileReport(df, title='EDA Report', explorative=True)\n",
    "profile.to_file('eda_report.html')  # Save as HTML\n",
    "profile.to_notebook_iframe()        # Display in Jupyter notebook\n",
    "```\n",
    "\n",
    "**Information Automatically Captured:**\n",
    "\n",
    "| Section | Contents |\n",
    "|---|---|\n",
    "| **Overview** | Dataset shape, memory size, number of variables, duplicates, missing cells total |\n",
    "| **Variables** | Per-column: type, unique values, missing count/%, distinct count, histogram, common values |\n",
    "| **Quantile Stats** | Min, Q1, Median, Q3, Max, IQR, Range |\n",
    "| **Descriptive Stats** | Mean, std, variance, skewness, kurtosis, coefficient of variation |\n",
    "| **Correlations** | Pearson, Spearman, Kendall, Cramér's V (for categoricals) heatmaps |\n",
    "| **Missing Values** | Heatmap, bar chart, and dendrogram of missing patterns |\n",
    "| **Interactions** | Scatter plot matrix for numerical variables |\n",
    "| **Sample** | First and last 10 rows of the dataset |\n",
    "| **Alerts** | Warnings for high correlation, high cardinality, imbalanced data, constant columns |\n",
    "\n",
    "**How It Speeds Up the Data Science Workflow:**\n",
    "\n",
    "1. **Instant Profiling:** Replaces 50–200 lines of manual EDA code with 3 lines — saving 30-60 minutes per dataset\n",
    "2. **Smart Alerts:** Automatically flags problem areas (e.g., \"Age has 12% missing values\", \"Income is highly skewed\") — prioritizes what needs attention\n",
    "3. **Shareable Reports:** The HTML output can be shared with stakeholders/domain experts without requiring Python knowledge\n",
    "4. **Correlation Discovery:** Automatic cross-correlation matrices reveal feature relationships that might be missed manually\n",
    "5. **Baseline Documentation:** Provides a reproducible, versioned snapshot of dataset characteristics at each pipeline stage"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q29. Python: Automate outlier detection using IQR and Z-score methods. Compare and visualize with side-by-side box plots. [10 marks]"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import numpy as np\n",
    "import pandas as pd\n",
    "import matplotlib.pyplot as plt\n",
    "from scipy import stats\n",
    "\n",
    "def detect_outliers_iqr(data):\n",
    "    \"\"\"Detect outliers using IQR method.\"\"\"\n",
    "    Q1 = np.percentile(data, 25)\n",
    "    Q3 = np.percentile(data, 75)\n",
    "    IQR = Q3 - Q1\n",
    "    lower = Q1 - 1.5 * IQR\n",
    "    upper = Q3 + 1.5 * IQR\n",
    "    outliers = data[(data < lower) | (data > upper)]\n",
    "    clean = data[(data >= lower) & (data <= upper)]\n",
    "    return outliers, clean, lower, upper\n",
    "\n",
    "def detect_outliers_zscore(data, threshold=3.0):\n",
    "    \"\"\"Detect outliers using Z-score method.\"\"\"\n",
    "    z_scores = np.abs(stats.zscore(data))\n",
    "    outliers = data[z_scores > threshold]\n",
    "    clean = data[z_scores <= threshold]\n",
    "    return outliers, clean\n",
    "\n",
    "def compare_outlier_methods(series, col_name='Feature'):\n",
    "    \"\"\"\n",
    "    Detect and compare outliers using IQR and Z-score methods.\n",
    "    Visualizes results with side-by-side box plots.\n",
    "    \"\"\"\n",
    "    data = series.dropna().values\n",
    "\n",
    "    iqr_out, iqr_clean, lb, ub = detect_outliers_iqr(data)\n",
    "    z_out, z_clean = detect_outliers_zscore(data)\n",
    "\n",
    "    print(f\"\\n{'=' * 55}\")\n",
    "    print(f\"OUTLIER COMPARISON — {col_name}\")\n",
    "    print(f\"{'=' * 55}\")\n",
    "    print(f\"  Total observations      : {len(data)}\")\n",
    "    print(f\"\\n  IQR Method:\")\n",
    "    print(f\"    Q1={np.percentile(data,25):.2f}, Q3={np.percentile(data,75):.2f}, IQR={np.percentile(data,75)-np.percentile(data,25):.2f}\")\n",
    "    print(f\"    Bounds: [{lb:.2f}, {ub:.2f}]\")\n",
    "    print(f\"    Outliers detected : {len(iqr_out)} ({len(iqr_out)/len(data)*100:.1f}%)\")\n",
    "    print(f\"    Outlier values    : {sorted(iqr_out)[:10]}\")\n",
    "\n",
    "    print(f\"\\n  Z-score Method (|z| > 3):\")\n",
    "    print(f\"    Mean={np.mean(data):.2f}, Std={np.std(data):.2f}\")\n",
    "    print(f\"    Outliers detected : {len(z_out)} ({len(z_out)/len(data)*100:.1f}%)\")\n",
    "    print(f\"    Outlier values    : {sorted(z_out)[:10]}\")\n",
    "\n",
    "    print(f\"\\n  IQR detects {len(iqr_out) - len(z_out):+d} more outliers than Z-score\")\n",
    "\n",
    "    # Visualization — Side-by-side box plots\n",
    "    fig, axes = plt.subplots(1, 3, figsize=(15, 6))\n",
    "    fig.suptitle(f'Outlier Detection Comparison — {col_name}', fontsize=14, fontweight='bold')\n",
    "\n",
    "    datasets = [\n",
    "        (data,      'Original Data',       '#3498db'),\n",
    "        (iqr_clean, 'After IQR Removal',   '#2ecc71'),\n",
    "        (z_clean,   'After Z-score Removal','#e74c3c')\n",
    "    ]\n",
    "\n",
    "    for ax, (d, title, color) in zip(axes, datasets):\n",
    "        bp = ax.boxplot(d, patch_artist=True,\n",
    "                        boxprops=dict(facecolor=color, alpha=0.6),\n",
    "                        medianprops=dict(color='black', linewidth=2),\n",
    "                        flierprops=dict(marker='o', markerfacecolor='orange',\n",
    "                                        markersize=5, alpha=0.7))\n",
    "        ax.set_title(f'{title}\\n(n={len(d)})', fontweight='bold')\n",
    "        ax.set_ylabel(col_name)\n",
    "        stats_text = f'Mean: {np.mean(d):.1f}\\nStd: {np.std(d):.1f}\\nMin: {np.min(d):.1f}\\nMax: {np.max(d):.1f}'\n",
    "        ax.text(1.35, np.percentile(d, 75), stats_text, fontsize=8,\n",
    "                verticalalignment='top',\n",
    "                bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))\n",
    "\n",
    "    plt.tight_layout()\n",
    "    plt.savefig(f'outlier_comparison_{col_name.lower().replace(\" \",\"_\")}.png', dpi=150)\n",
    "    plt.show()\n",
    "\n",
    "\n",
    "# ── Demo ──────────────────────────────────────────────────────────\n",
    "np.random.seed(42)\n",
    "# Simulated income data (right-skewed with injected outliers)\n",
    "income = np.concatenate([\n",
    "    np.random.normal(50000, 12000, 200),\n",
    "    [200000, 250000, 350000, 10000, 5000, -2000, 500000]  # injected outliers\n",
    "])\n",
    "\n",
    "income_series = pd.Series(income, name='Income')\n",
    "compare_outlier_methods(income_series, col_name='Income')"
   ]
  },
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "### Q30. Python: End-to-end automated EDA pipeline — load/profile, handle missing values, encode categoricals, scale features, output summary and correlation heatmap. [10 marks]"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import numpy as np\n",
    "import pandas as pd\n",
    "import matplotlib.pyplot as plt\n",
    "import seaborn as sns\n",
    "import os\n",
    "\n",
    "from sklearn.preprocessing import LabelEncoder, StandardScaler\n",
    "from sklearn.impute import SimpleImputer\n",
    "\n",
    "# ── STEP 0: Generate a sample CSV dataset ─────────────────────────\n",
    "np.random.seed(42)\n",
    "n = 300\n",
    "raw_df = pd.DataFrame({\n",
    "    'Age':        np.random.randint(18, 70, n).astype(float),\n",
    "    'Income':     np.random.exponential(50000, n),\n",
    "    'Education':  np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], n),\n",
    "    'Experience': np.random.randint(0, 40, n).astype(float),\n",
    "    'Score':      np.random.normal(65, 15, n),\n",
    "    'Employed':   np.random.choice(['Yes', 'No'], n),\n",
    "    'Region':     np.random.choice(['North', 'South', 'East', 'West'], n)\n",
    "})\n",
    "# Inject missing values\n",
    "for col, frac in [('Age', 0.05), ('Income', 0.08), ('Education', 0.06), ('Score', 0.1)]:\n",
    "    idx = np.random.choice(n, int(n * frac), replace=False)\n",
    "    raw_df.loc[idx, col] = np.nan\n",
    "\n",
    "raw_df.to_csv('sample_data.csv', index=False)\n",
    "print(\"✅ sample_data.csv generated.\")\n",
    "\n",
    "\n",
    "# ═══════════════════════════════════════════════════════════════════\n",
    "def automated_eda_pipeline(csv_path, output_dir='pipeline_outputs'):\n",
    "    \"\"\"\n",
    "    End-to-end automated EDA pipeline.\n",
    "    Steps: Load → Profile → Handle Missing → Encode → Scale → Report\n",
    "    \"\"\"\n",
    "    os.makedirs(output_dir, exist_ok=True)\n",
    "    SEP = \"═\" * 60\n",
    "\n",
    "    # ── STEP 1: Load & Profile ─────────────────────────────────────\n",
    "    print(SEP)\n",
    "    print(\"  STEP 1: LOAD & PROFILE\")\n",
    "    print(SEP)\n",
    "    df = pd.read_csv(csv_path)\n",
    "    print(f\"  Shape       : {df.shape}\")\n",
    "    print(f\"  Memory      : {df.memory_usage(deep=True).sum() / 1024:.1f} KB\")\n",
    "    print(f\"  Duplicates  : {df.duplicated().sum()}\")\n",
    "    print(\"\\n  Dtypes:\")\n",
    "    print(df.dtypes.to_string())\n",
    "\n",
    "    num_cols = df.select_dtypes(include=np.number).columns.tolist()\n",
    "    cat_cols = df.select_dtypes(include='object').columns.tolist()\n",
    "    print(f\"\\n  Numerical : {num_cols}\")\n",
    "    print(f\"  Categorical: {cat_cols}\")\n",
    "\n",
    "    # ── STEP 2: Handle Missing Values ──────────────────────────────\n",
    "    print(f\"\\n{SEP}\")\n",
    "    print(\"  STEP 2: HANDLE MISSING VALUES\")\n",
    "    print(SEP)\n",
    "    print(\"  Before:\")\n",
    "    null_info = df.isnull().sum()\n",
    "    print(null_info[null_info > 0].to_string())\n",
    "\n",
    "    # Numerical: median imputation\n",
    "    num_imputer = SimpleImputer(strategy='median')\n",
    "    df[num_cols] = num_imputer.fit_transform(df[num_cols])\n",
    "\n",
    "    # Categorical: most frequent imputation\n",
    "    cat_imputer = SimpleImputer(strategy='most_frequent')\n",
    "    df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])\n",
    "\n",
    "    print(f\"\\n  After imputation — Missing values remaining: {df.isnull().sum().sum()}\")\n",
    "\n",
    "    # ── STEP 3: Encode Categorical Variables ───────────────────────\n",
    "    print(f\"\\n{SEP}\")\n",
    "    print(\"  STEP 3: ENCODE CATEGORICAL VARIABLES\")\n",
    "    print(SEP)\n",
    "    le = LabelEncoder()\n",
    "    encoding_map = {}\n",
    "    for col in cat_cols:\n",
    "        df[col + '_enc'] = le.fit_transform(df[col])\n",
    "        encoding_map[col] = dict(zip(le.classes_, le.transform(le.classes_)))\n",
    "        print(f\"  {col}: {encoding_map[col]}\")\n",
    "\n",
    "    # Drop original categorical columns\n",
    "    df.drop(columns=cat_cols, inplace=True)\n",
    "    enc_cols = [c + '_enc' for c in cat_cols]\n",
    "\n",
    "    # ── STEP 4: Scale Numerical Features ──────────────────────────\n",
    "    print(f\"\\n{SEP}\")\n",
    "    print(\"  STEP 4: SCALE NUMERICAL FEATURES (StandardScaler)\")\n",
    "    print(SEP)\n",
    "    scaler = StandardScaler()\n",
    "    df_scaled = df.copy()\n",
    "    df_scaled[num_cols] = scaler.fit_transform(df[num_cols])\n",
    "\n",
    "    print(\"  Scaling applied. Sample statistics after scaling:\")\n",
    "    print(df_scaled[num_cols].describe().loc[['mean', 'std']].round(3).to_string())\n",
    "\n",
    "    # ── STEP 5: Summary Report & Correlation Heatmap ──────────────\n",
    "    print(f\"\\n{SEP}\")\n",
    "    print(\"  STEP 5: SUMMARY REPORT & VISUALIZATION\")\n",
    "    print(SEP)\n",
    "\n",
    "    # Save summary CSV\n",
    "    summary = df_scaled.describe().T\n",
    "    summary_path = os.path.join(output_dir, 'pipeline_summary.csv')\n",
    "    summary.to_csv(summary_path)\n",
    "    print(f\"  💾 Summary saved: {summary_path}\")\n",
    "\n",
    "    # Correlation heatmap\n",
    "    corr = df_scaled.corr()\n",
    "    fig, ax = plt.subplots(figsize=(10, 8))\n",
    "    mask = np.triu(np.ones_like(corr, dtype=bool))\n",
    "    sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,\n",
    "                mask=mask, linewidths=0.4, ax=ax,\n",
    "                cbar_kws={'shrink': 0.7})\n",
    "    ax.set_title('Correlation Heatmap — Processed Dataset',\n",
    "                 fontsize=13, fontweight='bold')\n",
    "    plt.tight_layout()\n",
    "    heatmap_path = os.path.join(output_dir, 'pipeline_correlation_heatmap.png')\n",
    "    plt.savefig(heatmap_path, dpi=150)\n",
    "    plt.show()\n",
    "    print(f\"  💾 Heatmap saved: {heatmap_path}\")\n",
    "\n",
    "    print(f\"\\n{SEP}\")\n",
    "    print(f\"  ✅ PIPELINE COMPLETE — Outputs in: ./{output_dir}/\")\n",
    "    print(SEP)\n",
    "\n",
    "    return df_scaled\n",
    "\n",
    "\n",
    "# ── Run the pipeline ──────────────────────────────────────────────\n",
    "processed_df = automated_eda_pipeline('sample_data.csv', output_dir='pipeline_outputs')\n",
    "print(f\"\\nProcessed DataFrame shape: {processed_df.shape}\")\n",
    "processed_df.head()"
   ]
  }
 ],
 "metadata": {
  "kernelspec": {
   "display_name": "Python 3",
   "language": "python",
   "name": "python3"
  },
  "language_info": {
   "name": "python",
   "version": "3.10.0"
  }
 },
 "nbformat": 4,
 "nbformat_minor": 5
}
````
